## Importando dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.linear_model import LinearRegression
from scipy.stats import gaussian_kde as kde
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("Dados_Radar_Estacao_Completo_2018_2022.csv", index_col=0, parse_dates=['time'])
```

## Separando Treino e Teste

```python
plt.style.use('ggplot')
```

```python
test = df[df['time'].dt.year == 2022]
```

```python
train = df[df['time'].dt.year != 2022]
```

```python
train = train[train['Tp_est'] != 0]
```

```python
test = test[test['Tp_est'] != 0]
```

```python
plt.figure(figsize=(12, 8))
train['time'].dt.year.value_counts().plot(kind='pie')
plt.show()
```

## Tratando dados

```python
# Dados não nulos
for column in train.columns:
    print(column, train[column].isna().value_counts().loc[False])
```

```python
train = train.dropna()
test = test.dropna()
```

```python
print("tamanho do treino = ", len(train), "\ntamanho do teste = ", len(test))
```

### Aumentando a linearidade dos meses
#### A ideia é que o mês de dezembro deva ter a mesma distância pra novembro e pra janeiro

```python
months = [[], []]
for i in range(1,13):
    x = math.cos(math.radians(i*30))
    y = math.sin(math.radians(i*30))
    months[0].append(x)
    months[1].append(y)
```

#### Dados já ficam normalizados entre -1 e 1

```python
plt.plot(months[0], months[1])
plt.show()
```

#### Podemos calcular as distâncias pra confirmar funciona como esperado

```python
def dist(a1, b1, a2, b2):
    d1 = max(a1, a2) - min(a1, a2)
    d2 = max(b1, b2) - min(b1, b2)
    d = d1 ** 2 + d2 ** 2
    return math.sqrt(d)
```

```python
for i in range(12):
    print(dist(months[0][0], months[1][0], months[0][i], months[1][i]))
```

#### Agora transformamos os meses nessas duas colunas

```python
def converte_meses_x(series):
    m = series['time'].month
    return math.cos(math.radians(m*30))
```

```python
def converte_meses_y(series):
    m = series['time'].month
    return math.sin(math.radians(m*30))
```

```python
train['month_x'] = train.apply(converte_meses_x, axis=1)
train['month_y'] = train.apply(converte_meses_y, axis=1)
test['month_x'] = test.apply(converte_meses_x, axis=1)
test['month_y'] = test.apply(converte_meses_y, axis=1)
```

#### Dropamos o tempo original visto que já extraímos a informação relevante dele

```python
train.drop('time', axis=1, inplace=True)
test.drop('time', axis=1, inplace=True)
```

### Transformando as estações de coleta em inteiros

```python
est_dict = {}
i = 0
for est in np.sort(train['Est'].unique()):
    est_dict[est] = i
    i += 1
```

```python
est_dict
```

```python
def converte_estacoes(series, est_dict):
    if series['Est'] in est_dict:
        return est_dict[series['Est']]
    return None
```

```python
train['Est'] = train.apply(converte_estacoes, est_dict=est_dict, axis=1)
test['Est'] = test.apply(converte_estacoes, est_dict=est_dict, axis=1)
test.dropna(inplace=True)
```

```python
train.reset_index(drop=True, inplace=True)
test.reset_index(drop=True, inplace=True)
```

### Normalizando dados

```python
train_x = train.drop('Tp_est', axis=1)
train_y = train['Tp_est']
test_x = test.drop('Tp_est', axis=1)
test_y = test['Tp_est']
```

```python

scaler = MinMaxScaler()
scaler.fit(train_x)
train_x = scaler.transform(train_x)
test_x = scaler.transform(test_x)
```

### Testando Linear Regression
#### Como foi o melhor nos testes, deixei como resultado e comentei os outros

```python
reg = LinearRegression()
reg.fit(train_x, train_y)
y_pred = reg.predict(test_x)
```

```python
samples = test_y, y_pred
densObj = kde(samples)
plt.figure(figsize=(10,8))
plt.scatter(x=test_y, y=y_pred, c=densObj.evaluate(samples), cmap='gnuplot')
plt.title("Gráfico de dispersão")
plt.xlabel("Tp_est real")
plt.ylabel("Tp_est prevista")
plt.show()
print("Mean Square Error = ", mean_squared_error(y_true=test_y, y_pred=y_pred))
print("Mean Absolute Error = ", mean_absolute_error(y_true=test_y, y_pred=y_pred))
```
